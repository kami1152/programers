```python
def solution(s):
  L_cnt = 0
  R_cnt = 0
  count = len(s)
  
  if s[0] == ")":
    return False

  for i in range(count):
    if s[i] == "(":
      L_cnt += 1
    elif s[i] == ")":
      R_cnt += 1

  if R_cnt != L_cnt:
    return False

  for i in range(count):
    if s[i] == "(":
      L_cnt -= 1
    elif s[i] == ")":
      R_cnt -= 1

    if L_cnt > R_cnt:
      return False

  return True

a = "()())"
solution(a)
```

```python
import math
def fact(n):
  if(n>1):
    return n*fact(n-1)
  else:
    return 1

def ASP(q, z): #q = 0.1 z = int
  p = 1.0 -q
  lam = z*q/p
  sum = 1.0
  for k in range(z):
    ret = (lam**k)*(math.e**(-lam))/fact(k)*(1-(q/p)**(z-k))
    sum -= ret
  return sum

for i in range(0, 51, 5):
  print(f'z={i},  P={ASP(0.3, i)}')

```
